```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("train (1).csv")
test = pd.read_csv("test (1).csv")
```

```python
#1 - 2
train.describe()
```

```python
train['is_internet']=train['four_g']+train['three_g']+train['wifi']
train['is_internet']=train['is_internet'].replace(2, 1)
train['is_internet']=train['is_internet'].replace(3, 1)
train.head()
```

```python
#2 - 0.12

df_train = train [(train['is_internet']==0)]
len(df_train)/len(train)
```

```python
test['is_internet']=test['four_g']+test['three_g']+test['wifi']
test['is_internet']=test['is_internet'].replace(2, 1)
test['is_internet']=test['is_internet'].replace(3, 1)
test.head()
```

```python
#3a - 0.92

corr = train.corr(numeric_only=True)
corr
```

```python
#3b - 0.52

df_train = train [(train['price_range']==2) & (train['blue']==0)]
len(df_train)/len(train[(train['price_range']==2)])
```

```python
#3c - 48

print(len(train[(train['dual_sim']==1) & (train["talk_time"]<10)]))
print(len(train[(train['dual_sim']==0) & (train["talk_time"]<10)]))
```

```python
#4a - 3

train.nunique()
```

```python
train['touch_screen'].value_counts()
```

```python
train['touch_screen']=train['touch_screen'].replace('-', 'Yes')
```

```python
train['touch_screen'].value_counts()
```

```python
test['touch_screen'].value_counts()
```

```python
test['touch_screen']=test['touch_screen'].replace('-', 'No')
```

```python
test['touch_screen'].value_counts()
```

```python
X = train.drop(columns='price_range')
y = train['price_range']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

ohe = OneHotEncoder(drop='first', sparse=False)

ohe.fit(X[['touch_screen']])

X_encoded = pd.DataFrame(ohe.transform(X[['touch_screen']]), columns=ohe.get_feature_names_out(['touch_screen']))
X = X.drop(columns=['touch_screen']).join(X_encoded)
```

```python
test_encoded = pd.DataFrame(ohe.transform(test[['touch_screen']]), columns=ohe.get_feature_names_out(['touch_screen']))
test = test.drop(columns=['touch_screen']).join(test_encoded)
```

```python
#4b - 0.63
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 42)

cross_val_score(model, X, y, scoring='f1_weighted', cv=3).mean()
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

gs = GridSearchCV(LogisticRegression(random_state = 42), params, scoring = 'f1_weighted', cv=3)
gs.fit(X,y)
```

```python
#5a - 1
gs.best_params_
```
